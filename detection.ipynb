```python
# Importing YOLO

from ultralytics import YOLO
```

```python
# Using Yolo version 8m on the first dataset

model = YOLO('yolov8m')

model.train(data = '/content/dataset/data.yaml',epochs=150, imgsz=416, batch=-1, patience=20)
```

```python
# Using the best weights the model achived to test the model

model = YOLO('models/signdetection.pt')

# predicting using Confidence Threshold is 20%, the Overlap Threshold is 20%, and its limited to 3 detections per frame/image

model.predict(source = 'data/test/images',conf=0.2, iou=0.2, save=True, max_det=1)
```
